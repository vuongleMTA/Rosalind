## 5. Computing GC Content

1. **XÃ¡c Ä‘á»‹nh ngÃ´n ngá»¯ tá»« vÄƒn báº£n** cÃ³ thá»ƒ thá»±c hiá»‡n báº±ng cÃ¡ch **phÃ¢n tÃ­ch táº§n suáº¥t chá»¯ cÃ¡i**, vÃ¬ má»—i ngÃ´n ngá»¯ cÃ³ Ä‘áº·c trÆ°ng riÃªng vá» táº§n suáº¥t chá»¯ cÃ¡i.
2. **XÃ¡c Ä‘á»‹nh loÃ i tá»« DNA** cÅ©ng cÃ³ thá»ƒ thá»±c hiá»‡n báº±ng cÃ¡ch **phÃ¢n tÃ­ch táº§n suáº¥t cÃ¡c base** (A, T, C, G).
3. **GC-content** (tá»· lá»‡ pháº§n trÄƒm cá»§a cytosine vÃ  guanine trong DNA) lÃ  má»™t chá»‰ sá»‘ quan trá»ng Ä‘á»ƒ **phÃ¢n biá»‡t loÃ i**.
4. **Sinh váº­t nhÃ¢n thá»±c (eukaryote)** thÆ°á»ng cÃ³ **GC-content khoáº£ng 50%**, trong khi **sinh váº­t nhÃ¢n sÆ¡ (prokaryote)** thÆ°á»ng cÃ³ **GC-content cao hÆ¡n 50%**.
5. GC-content cÃ³ thá»ƒ Ä‘Æ°á»£c sá»­ dá»¥ng Ä‘á»ƒ **nhanh chÃ³ng phÃ¢n loáº¡i sinh váº­t** ngay cáº£ tá»« má»™t máº«u DNA ráº¥t nhá».

GC-content lÃ  pháº§n trÄƒm sá»‘ lÆ°á»£ng base 'C' vÃ  'G' trong má»™t chuá»—i DNA.
Chuá»—i bá»• sung ngÆ°á»£c (reverse complement) cÃ³ cÃ¹ng GC-content vá»›i chuá»—i gá»‘c.
FASTA format Ä‘Æ°á»£c sá»­ dá»¥ng Ä‘á»ƒ lÆ°u trá»¯ chuá»—i DNA vá»›i nhÃ£n báº¯t Ä‘áº§u báº±ng dáº¥u '>' vÃ  cÃ¡c dÃ²ng tiáº¿p theo chá»©a chuá»—i DNA.
Má»¥c tiÃªu bÃ i toÃ¡n lÃ  tÃ¬m chuá»—i cÃ³ GC-content cao nháº¥t trong danh sÃ¡ch cÃ¡c chuá»—i FASTA Ä‘Æ°á»£c cho.

```python
def compu_GC(s):
    G = s.count('G')
    C = s.count('C')
    length = len(s)
    GC = (G+C)/length
    return GC
```

# Mendel'first Law

Báº¥t kÃ¬ 1 tÃ­nh tráº¡ng/yáº¿u tá»‘ nÃ o cÅ©ng táº¡o bá»Ÿi alen, cÃ³ thá»ƒ trá»™i or láº·n

### BÃ€I TOÃN

BÃ i toÃ¡n yÃªu cáº§u tÃ­nh xÃ¡c suáº¥t mÃ  hai sinh váº­t Ä‘Æ°á»£c chá»n ngáº«u nhiÃªn tá»« má»™t quáº§n thá»ƒ sáº½ sinh ra má»™t con cÃ¡i mang alen trá»™i (vÃ  do Ä‘Ã³ thá»ƒ hiá»‡n kiá»ƒu hÃ¬nh trá»™i). ChÃºng ta sáº½ lÃ m Ä‘iá»u nÃ y báº±ng cÃ¡ch sá»­ dá»¥ng xÃ¡c suáº¥t cá»§a cÃ¡c cáº·p sinh váº­t vá»›i kiá»ƒu gen khÃ¡c nhau.

### Äáº§u vÃ o:
- **k**: Sá»‘ lÆ°á»£ng cÃ¡ thá»ƒ Ä‘á»“ng há»£p trá»™i (homozygous dominant).
- **m**: Sá»‘ lÆ°á»£ng cÃ¡ thá»ƒ dá»‹ há»£p (heterozygous).
- **n**: Sá»‘ lÆ°á»£ng cÃ¡ thá»ƒ Ä‘á»“ng há»£p láº·n (homozygous recessive).

### TÃ­nh xÃ¡c suáº¥t chá»n má»—i loáº¡i cáº·p:

Tá»•ng sá»‘ cÃ¡ thá»ƒ lÃ  \( k + m + n \).  
XÃ¡c suáº¥t chá»n má»™t cáº·p cÃ¡ thá»ƒ dá»±a trÃªn kiá»ƒu gen cá»§a chÃºng sáº½ lÃ  tá»· lá»‡ giá»¯a sá»‘ lÆ°á»£ng cÃ¡ thá»ƒ kiá»ƒu gen Ä‘Ã³ vÃ  tá»•ng sá»‘ cÃ¡ thá»ƒ trong quáº§n thá»ƒ.

### XÃ¡c suáº¥t cá»§a tá»«ng cáº·p:
1. **Cáº·p AA-AA**:  
   XÃ¡c suáº¥t lÃ :  
   {k}{k+m+n} \times \frac{k-1}{k+m+n-1}
   

2. **Cáº·p AA-Aa**:  
   XÃ¡c suáº¥t lÃ :  
   \[
   \frac{k}{k+m+n} \times \frac{m}{k+m+n-1}
   \]

3. **Cáº·p AA-aa**:  
   XÃ¡c suáº¥t lÃ :  
   \[
   \frac{k}{k+m+n} \times \frac{n}{k+m+n-1}
   \]

4. **Cáº·p Aa-Aa**:  
   XÃ¡c suáº¥t lÃ :  
   \[
   \frac{m}{k+m+n} \times \frac{m-1}{k+m+n-1}
   \]

5. **Cáº·p Aa-aa**:  
   XÃ¡c suáº¥t lÃ :  
   \[
   \frac{m}{k+m+n} \times \frac{n}{k+m+n-1}
   \]

6. **Cáº·p aa-aa**:  
   XÃ¡c suáº¥t lÃ :  
   \[
   \frac{n}{k+m+n} \times \frac{n-1}{k+m+n-1}
   \]

### CÃ´ng thá»©c tÃ­nh xÃ¡c suáº¥t tá»•ng thá»ƒ:
Tá»•ng xÃ¡c suáº¥t sinh ra con cÃ¡i mang alen trá»™i lÃ  tá»•ng cá»§a xÃ¡c suáº¥t sinh ra con cÃ¡i mang alen trá»™i tá»« táº¥t cáº£ cÃ¡c loáº¡i cáº·p:

\[
P(\text{dominant}) = P(\text{AA-AA}) + P(\text{AA-Aa}) + P(\text{AA-aa}) + P(\text{Aa-Aa}) + P(\text{Aa-aa})
\]


```python
def calculate_probability(k, m, n):
    total = k + m + n

    # TÃ­nh xÃ¡c suáº¥t cho má»—i loáº¡i cáº·p
    p_AA_AA = (k / total) * ((k - 1) / (total - 1))  # Cáº·p AA-AA
    p_AA_Aa = (k / total) * (m / (total - 1)) + ((m / total)*k/(total-1)) # Cáº·p AA-Aa
    p_AA_aa = (k / total) * (n / (total - 1)) + (n/total)*k/(total-1) # Cáº·p AA-aa
    p_Aa_Aa = (m / total) * ((m - 1) / (total - 1))  # Cáº·p Aa-Aa
    p_Aa_aa = (m / total) * (n / (total - 1))+(n/total)*m/(total-1)  # Cáº·p Aa-aa
    p_aa_aa = (n / total) * ((n - 1) / (total - 1))  # Cáº·p aa-aa

    # XÃ¡c suáº¥t tá»•ng há»£p sinh con cÃ¡i mang alen trá»™i
    p_dominant = (p_AA_AA * 1) + (p_AA_Aa * 1) + (p_AA_aa * 1) + (p_Aa_Aa * 0.75) + (p_Aa_aa * 0.5) + (p_aa_aa * 0)

    return p_dominant

# Äáº§u vÃ o
k, m, n = 2, 2, 2

# TÃ­nh xÃ¡c suáº¥t vÃ  in káº¿t quáº£
result = calculate_probability(k, m, n)
print(f"{result:.5f}")
```

# Finding a Most Likely Common Ancestor (TÃ¬m tá»• tiÃªn gáº§n nháº¥t)

```python
def max_charac(colum_i):
    a = colum_i.count('A')
    c = colum_i.count('C')
    g = colum_i.count('G')
    t = colum_i.count('T')
    num_dict = {}
    num_dict['A'] = a
    num_dict['C'] = c
    num_dict['G'] = g
    num_dict['T'] = t
    num_array =[]
    num_array.append(a)
    num_array.append(c)
    num_array.append(g)
    num_array.append(t)
    max = 0
    max_char = ""
    for item in num_dict.items():
        if item[1] > max:
            max = item[1]
            max_char = item[0]
    return max_char, num_array

colum_i = ['A', 'T', 'C', 'C', 'A', 'G', 'C', 'T']
max_charac(colum_i)
```

# Mortal Fibonacci Rabbits

```python
def mortal_fibonacci(n, m):
    # Táº¡o danh sÃ¡ch lÆ°u sá»‘ lÆ°á»£ng thá» tá»«ng thÃ¡ng
    rabbits = [0] * (n + 1)  # Táº¡o danh sÃ¡ch cÃ³ n+1 pháº§n tá»­
    rabbits[1] = 1  # ThÃ¡ng Ä‘áº§u tiÃªn cÃ³ 1 cáº·p thá»

    # Duyá»‡t tá»« thÃ¡ng 2 Ä‘áº¿n thÃ¡ng n
    for i in range(2, n + 1):
        if i <= m:
            rabbits[i] = rabbits[i - 1] + rabbits[i - 2]  # Thá» chÆ°a cháº¿t
        else:
            rabbits[i] = rabbits[i - 1] + rabbits[i - 2] - rabbits[i - m - 1]  # Trá»« thá» cháº¿t

    return rabbits[n]  # Káº¿t quáº£ lÃ  sá»‘ thá» á»Ÿ thÃ¡ng thá»© n

n = 3
m=6

# Gá»i hÃ m vÃ  in káº¿t quáº£
print(mortal_fibonacci(n, m))
```
