## 5. Computing GC Content

1. **Xác định ngôn ngữ từ văn bản** có thể thực hiện bằng cách **phân tích tần suất chữ cái**, vì mỗi ngôn ngữ có đặc trưng riêng về tần suất chữ cái.
2. **Xác định loài từ DNA** cũng có thể thực hiện bằng cách **phân tích tần suất các base** (A, T, C, G).
3. **GC-content** (tỷ lệ phần trăm của cytosine và guanine trong DNA) là một chỉ số quan trọng để **phân biệt loài**.
4. **Sinh vật nhân thực (eukaryote)** thường có **GC-content khoảng 50%**, trong khi **sinh vật nhân sơ (prokaryote)** thường có **GC-content cao hơn 50%**.
5. GC-content có thể được sử dụng để **nhanh chóng phân loại sinh vật** ngay cả từ một mẫu DNA rất nhỏ.

GC-content là phần trăm số lượng base 'C' và 'G' trong một chuỗi DNA.
Chuỗi bổ sung ngược (reverse complement) có cùng GC-content với chuỗi gốc.
FASTA format được sử dụng để lưu trữ chuỗi DNA với nhãn bắt đầu bằng dấu '>' và các dòng tiếp theo chứa chuỗi DNA.
Mục tiêu bài toán là tìm chuỗi có GC-content cao nhất trong danh sách các chuỗi FASTA được cho.

```python
def compu_GC(s):
    G = s.count('G')
    C = s.count('C')
    length = len(s)
    GC = (G+C)/length
    return GC
```

```python
def permutations(arr, current=[]):
    result = []  # Tạo một danh sách mới trong mỗi lần gọi hàm
    if not arr:
        result.append(current)  # Lưu hoán vị vào result
        return result  # Trả về danh sách chứa một hoán vị duy nhất
    
    for i in range(len(arr)):
        result += permutations(arr[:i] + arr[i+1:], current + [arr[i]])  # Gộp kết quả từ các lời gọi đệ quy
    
    return result  # Trả về danh sách chứa tất cả hoán vị

result = permutations([1, 2, 3])  # Gọi hàm và lưu kết quả
print(result)
```

# Mendel'first Law

Bất kì 1 tính trạng/yếu tố nào cũng tạo bởi alen, có thể trội or lặn

### BÀI TOÁN

Bài toán yêu cầu tính xác suất mà hai sinh vật được chọn ngẫu nhiên từ một quần thể sẽ sinh ra một con cái mang alen trội (và do đó thể hiện kiểu hình trội). Chúng ta sẽ làm điều này bằng cách sử dụng xác suất của các cặp sinh vật với kiểu gen khác nhau.

### Đầu vào:
- **k**: Số lượng cá thể đồng hợp trội (homozygous dominant).
- **m**: Số lượng cá thể dị hợp (heterozygous).
- **n**: Số lượng cá thể đồng hợp lặn (homozygous recessive).

### Tính xác suất chọn mỗi loại cặp:

Tổng số cá thể là \( k + m + n \).  
Xác suất chọn một cặp cá thể dựa trên kiểu gen của chúng sẽ là tỷ lệ giữa số lượng cá thể kiểu gen đó và tổng số cá thể trong quần thể.

### Xác suất của từng cặp:
1. **Cặp AA-AA**:  
   Xác suất là:  
   {k}{k+m+n} \times \frac{k-1}{k+m+n-1}
   

2. **Cặp AA-Aa**:  
   Xác suất là:  
   \[
   \frac{k}{k+m+n} \times \frac{m}{k+m+n-1}
   \]

3. **Cặp AA-aa**:  
   Xác suất là:  
   \[
   \frac{k}{k+m+n} \times \frac{n}{k+m+n-1}
   \]

4. **Cặp Aa-Aa**:  
   Xác suất là:  
   \[
   \frac{m}{k+m+n} \times \frac{m-1}{k+m+n-1}
   \]

5. **Cặp Aa-aa**:  
   Xác suất là:  
   \[
   \frac{m}{k+m+n} \times \frac{n}{k+m+n-1}
   \]

6. **Cặp aa-aa**:  
   Xác suất là:  
   \[
   \frac{n}{k+m+n} \times \frac{n-1}{k+m+n-1}
   \]

### Công thức tính xác suất tổng thể:
Tổng xác suất sinh ra con cái mang alen trội là tổng của xác suất sinh ra con cái mang alen trội từ tất cả các loại cặp:

\[
P(\text{dominant}) = P(\text{AA-AA}) + P(\text{AA-Aa}) + P(\text{AA-aa}) + P(\text{Aa-Aa}) + P(\text{Aa-aa})
\]


```python
def calculate_probability(k, m, n):
    total = k + m + n

    # Tính xác suất cho mỗi loại cặp
    p_AA_AA = (k / total) * ((k - 1) / (total - 1))  # Cặp AA-AA
    p_AA_Aa = (k / total) * (m / (total - 1)) + ((m / total)*k/(total-1)) # Cặp AA-Aa
    p_AA_aa = (k / total) * (n / (total - 1)) + (n/total)*k/(total-1) # Cặp AA-aa
    p_Aa_Aa = (m / total) * ((m - 1) / (total - 1))  # Cặp Aa-Aa
    p_Aa_aa = (m / total) * (n / (total - 1))+(n/total)*m/(total-1)  # Cặp Aa-aa
    p_aa_aa = (n / total) * ((n - 1) / (total - 1))  # Cặp aa-aa

    # Xác suất tổng hợp sinh con cái mang alen trội
    p_dominant = (p_AA_AA * 1) + (p_AA_Aa * 1) + (p_AA_aa * 1) + (p_Aa_Aa * 0.75) + (p_Aa_aa * 0.5) + (p_aa_aa * 0)

    return p_dominant

# Đầu vào
k, m, n = 2, 2, 2

# Tính xác suất và in kết quả
result = calculate_probability(k, m, n)
print(f"{result:.5f}")
```

```python
def max_charac(colum_i):
    a = colum_i.count('A')
    c = colum_i.count('C')
    g = colum_i.count('G')
    t = colum_i.count('T')
    num_dict = {}
    num_dict['A'] = a
    num_dict['C'] = c
    num_dict['G'] = g
    num_dict['T'] = t
    num_array =[]
    num_array.append(a)
    num_array.append(c)
    num_array.append(g)
    num_array.append(t)
    max = 0
    max_char = ""
    for item in num_dict.items():
        if item[1] > max:
            max = item[1]
            max_char = item[0]
    return max_char, num_array

colum_i = ['A', 'T', 'C', 'C', 'A', 'G', 'C', 'T']
max_charac(colum_i)
```

# Mortal Fibonacci Rabbits

```python
def mortal_fibonacci(n, m):
    # Tạo danh sách lưu số lượng thỏ từng tháng
    rabbits = [0] * (n + 1)  # Tạo danh sách có n+1 phần tử
    rabbits[1] = 1  # Tháng đầu tiên có 1 cặp thỏ

    # Duyệt từ tháng 2 đến tháng n
    for i in range(2, n + 1):
        if i <= m:
            rabbits[i] = rabbits[i - 1] + rabbits[i - 2]  # Thỏ chưa chết
        else:
            rabbits[i] = rabbits[i - 1] + rabbits[i - 2] - rabbits[i - m - 1]  # Trừ thỏ chết

    return rabbits[n]  # Kết quả là số thỏ ở tháng thứ n

n = 3
m=6

# Gọi hàm và in kết quả
print(mortal_fibonacci(n, m))
```
